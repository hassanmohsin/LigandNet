Get the gene names for the drugs in the Colleens_selleck_chem_results_partial_8-28-20v4 from the known interactions

```python
import pandas as pd
```

```python
colleen_df = pd.read_csv("data/colleen_selleck_chem_results_partial_8-28-20v4.csv")
colleen_df = pd.DataFrame.drop_duplicates(colleen_df)
colleen_df.head()
```

```python
known_interactions_df = pd.read_csv("data/interactions.csv", usecols=[1, 2])
known_interactions_df = pd.DataFrame.drop_duplicates(known_interactions_df)
known_interactions_df.drug = known_interactions_df.drug.apply(lambda x: x.lower())
known_interactions_df.head()
```

```python
merge_df = pd.merge(colleen_df, known_interactions_df, left_on='Drug', right_on='drug', how='inner')
merge_df.head()
```

Find the predicted protein names from the LigandNet predictions using gene/PreyGene names

```python
from glob import glob
import os
prediction_files = glob("predictions/*.csv")
```

```python
def pred(file):
    drug_name = os.path.split(file)[1][:-4]
    _df = pd.read_csv(file)
    _df['drug'] = pd.Series([drug_name]*len(_df))
    return _df
```

```python
predictions_df = pd.concat([pred(p) for p in prediction_files], ignore_index=True)
predictions_df.drug = predictions_df.drug.apply(lambda x: x.lower())
predictions_df.head()
```

```python
len(predictions_df)
```

```python
gordon_to_predictrions_df = pd.merge(gordon_df, predictions_df, left_on='PreyGene', right_on='gene_name', how='inner')
gordon_to_predictrions_df.head()
```

```python
gordon_to_predictrions_df.to_csv('gordon_to_predictions.csv', index=False)
```
